```python
import numpy as np
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[10,20,30,40,50]
plt.plot(x,y,marker="o",ms=7,mec="red",mfc="yellow",color="blue",linewidth="3",linestyle=":")
f={"family":"serif","color":"brown","size":25}
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Graph",fontdict=f,loc="center")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x=["A","B","C","D"]
y=[10,20,30,40]
plt.scatter(y,x,linestyle=":",marker="*")
f={"family":"serif","color":"brown","size":25}
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Graph",fontdict=f,loc="center")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x=["A","B","C","D"]
y=[10,20,30,40]
plt.bar(x,y,linestyle=":",linewidth="")
f={"family":"serif","color":"brown"}
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Graph",fontdict=f,loc="center")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x=["A","B","C","D"]
y=[10,20,30,40]
plt.barh(x,y,height=0.3)
f={"family":"serif","color":"brown"}
plt.title("Graph",fontdict=f,loc="center")
plt.show()
```

```python
y=[10,20,30,40]
cat=["A","B","C","D"]
e=[0,0.2,0,0]
plt.pie(y,labels=cat,explode=e,autopct="%1.1f",radius=1.5)
plt.legend(title="Stats",loc="upper right")
plt.show()
```

```python
cat=['comedy','action','romance','drama','sci-fi']
stars=[4,5,6,1,4]
plt.subplot(2,5,1)
plt.pie(stars,labels=cat)
plt.subplot(2,5,3)
plt.barh(cat,stars)
plt.subplot(2,5,5)
plt.bar(cat,stars)
plt.subplot(2,5,7)
plt.plot(cat,stars)
plt.subplot(2,5,9)
plt.scatter(cat,stars)
plt.show()
```
