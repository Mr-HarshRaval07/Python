```python
import numpy as np
a=np.arange(1,101)
a1=np.array(a)
print(np.ndim(a1))#Dimensions
print(np.shape(a1))#shape
print(a1) #1-D
r1=a1.reshape(2,50)
print(np.ndim(r1))#Dimensions
print(np.shape(r1))#shape
print(r1) #2-D
r2=a1.reshape(2,5,10)#reshaping
print(np.ndim(r2))#Dimensions
print(np.shape(r2))#shape
print(r2) #3-D
print(r2[1,0:,0])#Indexing and slicing
```

```python
l1=[[1,2,6],[9,6,4],[7,0,3]]
ar1=np.array(l1)
print(ar1)
s1=np.sort(ar1,axis=1)#sorting elements row wise
print(s1)
s2=np.sort(ar1,axis=0)#sorting elements column wise
print(s2)
w1=np.where(ar1%2==0) #finding element using where
print(w1)
as1=np.array_split(ar1,2)#spliting array into number of pieces
print(as1)
```
