```python
import numpy as np
import matplotlib.pyplot as plt
Scores= [[13, 10, 9, 33], 
[63, 46, 90, 42], 
[39, 76, 13, 29], 
[82, 9, 29, 78], 
[67, 61, 59, 36]]
Match_6= [41, 87, 72, 36, 92]
Batsman_6= [77, 83, 98, 95, 89]
Batsman_7= [92, 71, 52, 61, 53]
sc=np.array(Scores)
m6=np.array(Match_6)
b6=np.array(Batsman_6).reshape(1,5)
b7=np.array(Batsman_7).reshape(1,5)
n1=m6.reshape(5,1)
updated_scores=np.concatenate((sc,n1),axis=1)
new_scores=np.concatenate((updated_scores,b6,b7),axis=0)
summ=np.sum(new_scores,axis=1).reshape(7,1)
final=np.concatenate((new_scores,summ),axis=1)
print(final)
```

```python
max_run_per_player=np.max(new_scores,axis=1)
x=[1,2,3,4,5,6,7]
plt.bar(x,max_run_per_player)
plt.show()
```

```python
max_run_per_match=np.max(new_scores,axis=0)
x=[1,2,3,4,5]
plt.plot(x,max_run_per_match)
plt.show()
```
