# Clone and copy

```python
x=[1,2,3]
z=x.copy()
print(x)
print(z)
x[0]=10
print(x)
print(z)
```

```python
x=[1,2,3,[4,5,6]]
z=x.copy()
x[0]=200
print(x)
print(z)
a=[[1,2,3],[4,5,6]]
b=a.copy()
a[0][2]=200
print(a)
print(b)
```

# List comprehension

```python
l=[1,2,3,4]
z=[i**2 if i%2==0 else i**3 for i in l]
print(z)
```

```python
x=[]
o=[]
e=[]
c=0
while c<10:
    n=int(input("Enter a number:"))
    c+=1
    x.append(n)
    if n%2==0:
        e.append(n)
    else:
        o.append(n)
x.sort()
o.sort()
e.sort()
print("List is :",x)
print("Odd list is :",o)
print("Even list is :",e)
print("Max number in main list is :",x[-1])
print("Min number in main list is :",x[0])
print("Second Max number in main list is :",x[-2])
print("Second Min number in main list is :",x[1])
print("Max number in odd list is :",o[-1])
print("Min number in odd list is :",o[0])
print("Max number in even list is :",e[-1])
print("Min number in even list is :",e[0])
```

# Codes

```python
x=['pyt','pytho','python','pytha']
x.sort(reverse=True)
ans=""
for i in range(len(x[-1])):
    if x[0][i]==x[-1][i]:
        ans=ans+x[-1][i]
        continue
    else:
        break
print(ans)
```

```python
l=[5,5,2,5,8]
for i in range(len(l)):
    x=l.copy()
    z=x.pop(i)
    e=sum(x[0::2])
    o=sum(x[1::2])
    if e==o:
        print("Removed ",z)
```
