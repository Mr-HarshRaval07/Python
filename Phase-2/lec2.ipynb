```python
u,l,d,s,space,w=0,0,0,0,0,0
while(w<3):
    p=input("Enter your password = ")
    
    if len(p)<8:
        print("Must contain 8 digits")
        w=w+1
    else:
        for i in p:
            if i.isupper():
                u=u+1
            elif i.islower():
                l=l+1
            elif i.isdigit():
                d=d+1
            elif i.isspace():
                space=space+1
            else:
                s=s+1
        if u>0 and l>0 and d>0 and space==0 and s>0:
            print("Valid")
        else:
            print("Invalid")
            w=w+1
        if w==3:
            print("blocked")
```

```python
final=""
word=(input("Enter pass = "))
enc=int(input("Enter encryption key"))
word=word.upper()
ans="ABCDEFGHIJKLMNOPQRSTUVWXYZ"
for i  in word:
    if i in ans:
        op=(ans.find(i)+enc)%26
        final=final+ans[op]
    else:
        final=final+i
print(final)
```
