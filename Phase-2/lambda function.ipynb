# Functions
Map()
filter()
reduce()

```python
l=[1,2,3,4]
z=list(map(lambda x:x+10,l))
print(z)
```

```python
z=list(filter(lambda x:x%2==0,l))
print(z)
```

```python
from functools import *
z=reduce(lambda x,y:x*y,l)
print(z)
```

# maketrans


```python
s="hello good morning"
ans=str.maketrans("ohg","@#9")
sol=s.translate(ans)
print(sol)
print(ans)
```

```python
s="he@#. INDIA is =,921"
ans=str.maketrans("I","i","#@=,.")
sol=s.translate(ans)
print(sol)
```
