```python
with open("HARDCODE.txt") as hc:
    x=hc.readlines()
    print(x)
    d={}
    for i in x:
        s=i.split()
        d[i]=int(s[1])
    print(d)
    ma=max(d.values())
    mi=min(d.values())
    for i,j in d.items():
        if j==ma:
            print(i,j)
        if j==mi:
            print(i,j)       
```
