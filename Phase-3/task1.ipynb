```python
with open("mbox-short.txt") as f:
    x=f.readlines()
    n=0
    l=[]
    for i in x:
        if "From " in i:
            n=n+1
            e=i.split()
            l.append(e[1])
print("No of email is ",n)
```

```python
d={}
for j in l:
    d[j]=l.count(j)
print(d)
m=max(d.values())
for i,j in d.items():
    if j==m:
        print(i,j)
```
