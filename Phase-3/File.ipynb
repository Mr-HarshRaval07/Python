```python
f1=open("f1.txt","w+")
f1.write("This is file 1 \n line 2 in file 1")
f1.seek(0)
x=f1.read()
f1.close()
f2=open("f2.txt","w+")
f2.write("This is file 2 \n line 2 in file 2")
f2.seek(0)
y=f2.read()
f2.close()
f3=open("f3.txt","w")
f3.write(x)
f3.write(y)
f3.close()
```
