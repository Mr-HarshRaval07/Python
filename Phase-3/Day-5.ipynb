```python
class Atm:
    def __init__(self):
       
        self.pin=0
        self.balance=0
        self.menu()
        
    def menu(self):
        print("1. check balance")
        print("2. withdraw money")
        print("3. deposit money")
        print("4. create pin")
        print("5. change pin")
        
        x = int(input("Enter choice"))
        
        if(x==1):
            self.CheckBalance()
        elif(x==2):
            self.WithdrawMethod()
        elif(x==3):
            self.DepositMethod()
        elif(x==4):
            self.CreatePin()
        elif(x==5):
            self.ChangePin()
    def CreatePin(self):
        self.pin = int(input("Enter pin"))
        if(len(str(self.pin))!=4):
            print("Invalid Pin")
            self.pin=0
            self.menu()
        
        self.balance = int(input("Enter balanced"))
        self.menu()
    def ChangePin(self):
        if(self.pin==0):
            print("first create pin")
            self.menu()
        y =int(input("Enter pin"))
        if(len(str(self.pin))!=4):
            print("Invalid Pin")
            self.menu()
        else:
            self.pin=y
            self.menu()
    def CheckBalance(self):
        y =int(input("Enter pin"))
        if(y==self.pin):
            print(self.balance)
            self.menu()
        else:
            print("Invalid Pin")
            self.menu()
    def WithdrawMethod(self):
        y =int(input("Enter pin"))
        if(y==self.pin):
            m = int(input("Enter Money to Withdraw"))
            if(m>self.balance):
                print("Invalid Money Input")
                self.menu()
            else:
                self.balance=self.balance-m
                print(self.balance)
                self.menu()
        else:
            print("Invalid Pin")
            self.menu()
    def DepositMethod(self):
        y =int(input("Enter pin"))
        if(y==self.pin):
            m = int(input("Enter Money to Deposit"))
            self.balance = self.balance + m
            print(self.balance)
            self.menu()
        else:
            print("Invalid Pin")
            self.menu()
            
x = Atm()
```
