```python
f = open("mbox-short.txt","r")
d={}

a="From "
l=[]
for i in f:
    if a in i:
        l1 = i.split()
        l.append(l1[1])
        
for i in l:
    d[i]=l.count(i)
# print(d)

ans = list(sorted(d.items(),key=lambda x:x[1],reverse=True))
print(ans[0][0],ans[0][1])
```
